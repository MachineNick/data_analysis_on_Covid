```python
covid_data_df = pd.read_csv('covid-countries-data.csv')
```

```python
combined_df = pd.merge(countries_df,covid_data_df,on='location')
```

```python
combined_df
```

```python
#  Q10: Add columns tests_per_million, cases_per_million and deaths_per_million into combined_df.
combined_df['tests_per_million'] = combined_df['total_tests'] * 1e6 / combined_df['population']
```

```python
combined_df['tests_per_million'] 
```

```python
combined_df['cases_per_million']  = combined_df['total_deaths']*1e6/ combined_df['population']
```

```python
#Create a dataframe with 10 countires that have highest number of tests per million people.
combined_df.sort_values(by=['cases_per_million'],ascending=True).head(10)
```
